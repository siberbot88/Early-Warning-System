**Membaca setiap file csv**

```python
import pandas as pd

courses = pd.read_csv(raw_path + "courses.csv")
assessments = pd.read_csv(raw_path + "assessments.csv")
vle = pd.read_csv(raw_path + "vle.csv")
studentInfo = pd.read_csv(raw_path + "studentInfo.csv")
studentRegistration = pd.read_csv(raw_path + "studentRegistration.csv")
studentAssessment = pd.read_csv(raw_path + "studentAssessment.csv")
studentVle = pd.read_csv(raw_path + "studentVle.csv")
```

**Cek tabel data**

```python
from IPython.display import display, Markdown

datasets = {
    "courses": courses,
    "assessments": assessments,
    "vle": vle,
    "studentInfo": studentInfo,
    "studentRegistration": studentRegistration,
    "studentAssessment": studentAssessment,
    "studentVle": studentVle
}

summary = []

for name, df in datasets.items():
    summary.append({
        "dataset": name,
        "rows": df.shape[0],
        "columns": df.shape[1]
    })

summary_df = pd.DataFrame(summary)
display(summary_df)

for name, df in datasets.items():
    display(Markdown(f"## Dataset: `{name}`"))
    display(Markdown(f"- **Ukuran**: {df.shape[0]} baris x {df.shape[1]} kolom"))
    display(Markdown(f"- **Kolom**: {', '.join(df.columns)}"))
    display(df.head())
```

**Cek Kolom**

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.columns.tolist())
```

**Cek Missing Values**

```python
for name, df in datasets.items():
    print(f"\nMissing values - {name}")
    print(df.isnull().sum())
```

**Ringkasan Missing Values**

```python
missing_tables = []

for dataset_name, df in datasets.items():
    missing = df.isnull().sum()
    missing = missing[missing > 0]

    for column_name, missing_count in missing.items():
        missing_tables.append({
            "Dataset": dataset_name,
            "Kolom": column_name,
            "Jumlah Missing Values": int(missing_count)
        })

missing_df = pd.DataFrame(missing_tables)
missing_df = missing_df.sort_values(by=["Dataset", "Jumlah Missing Values"], ascending=[True, False]).reset_index(drop=True)

display(missing_df)
```
